### 1. Find the Largest of Two Numbers
**Task:** Given two numbers, find the larger one.  
**Demo Input:**  5 and 6 <br>
**Output:** 6


```python
number1 = int(input("First number: "))
number2 = int(input("Second number: "))

if number1<number2:
    print(f"{number2} is larger than {number1}.")
elif number1>number2:
    print(f"{number1} is larger than {number2}.")
else:
    print(f"{number1} and {number2} both are equal.")
```

### 2. Print Numbers from 1 to N
**Task**: Given a number `n`, print all numbers from 1 to `n`. <br> 
**Demo Input**: `5`  <br>
**Demo Output**:  1 2 3 4 5


```python
number = int(input("How long do you want to print: "))
for i in range(1, number):
    print(i, end = ", ")
print(number)
```

### 3. Check if a Number is Positive or Negative
**Task**: Determine if the input number is positive, negative, or zero.

```python
number = int(input("Give me the number: "))
if number>0:
    print(f"The number {number} is Positive.")
elif number<0:
    print(f"The number {number} is Negative.")
else:
    print(f"The number {number} is Zero.")
```

### 4. Calculate the Sum of Digits
**Task**: Given a number, find the sum of its digits.  
**Demo Input**: `123`  
**Demo Output**: `6`  
(Explanation: `1+2+3=6`)

```python
number = input("Give me the number: ")
sum = 0
for i in number:
    sum += int(i)
print(sum)
```

### 5. Find Factorial of a Number
**Task**: Calculate the factorial of a given number.  
**Demo Input**: `4`  
**Demo Output**: `24`  
(Explanation: `4! = 4*3*2*1 = 24`)

```python
number = int(input("Which number: "))
factorial = 1
for i in range(1, number+1):
    factorial *= i
print(factorial)
```

### 6. Count Occurrences of a Digit
**Task**: Count how many times a specific digit appears max in a number.  
**Demo Input**:  1233321 <br>
**Demo Output**: `3`  <br>
(Explanation: `3` appears 3 times in `1233321`)

```python
number = input("Give your number: ")
max_num = 0
for digit in set(number):
    count = number.count(digit)
    if(count>max_num):
        max_num = count
        big_digit = digit
print(big_digit)
```

### 7. Find the GCD of Two Numbers
**Task**: Find the greatest common divisor (GCD) of two numbers.  
**Demo Input**: 8 and 12 <br>
**Demo Output**: `4`

```python
num1, num2 = map(int, input("Give me two numbers: ").split())
if num1 == 0:
    gcd = num1
elif num2 == 0:
    gcd = num2
else:
    while num2 != 0:
        t = num2
        num2 = num1 % num2
        num1 = t
    gcd = num1
print(f"GCD is {gcd}")
```

### 9. Check Armstrong Number
**Task**: Check if a number is an Armstrong number (the sum of its digits raised to the power of the number of digits equals the number).  
**Demo Input**: `153`  
**Demo Output**: `153 is an Armstrong Number`  
(Explanation: \(1^3 + 5^3 + 3^3 = 153\))

```python
num = int(input("Enter a number: "))
temp = num
digits = len(str(num))
total = 0
while temp > 0:
    d = temp % 10
    total += d ** digits
    temp //= 10

if total == num:
    print(f"{num} It's a Armstrong number.")
else:
    print(f"{num} It's not a Armstrong number")
```

### 10. Generate a Pattern
**Task**: Print a pyramid pattern with `n` rows.  
**Demo input:** 4 <br>
**Output:** You will see the pyramid pattern with 4 rows

```python
rows = int(input("How many rows would you like to see: "))
for i in range(1, rows+1):
    spaces = rows - i
    hashes = (2 * i) - 1
    print(" " * spaces + "#" * hashes)
```
